```python
import numpy as np
from sklearn.model_selection import train_test_split



#CASE 2
#When you have one input corresponding to one output. Jese y=sin(x). On value of x gives one value of y. Is case mein aik input layer aye gi
#############################################################################################################################################

import numpy as np
from sklearn.model_selection import train_test_split

# Define the input dataset and corresponding target variable
X = np.array([[0], [1]])
y = np.array([0, 1])

# Split the dataset into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Define the architecture of the neural network
input_layer_size = 1
hidden_layer_size = 2
output_layer_size = 1

# Initialize the weights and biases for each layer randomly
W1 = np.random.randn(input_layer_size, hidden_layer_size)
b1 = np.zeros((1, hidden_layer_size))
W2 = np.random.randn(hidden_layer_size, output_layer_size)
b2 = np.zeros((1, output_layer_size))

# Set the learning rate and the number of epochs
learning_rate = 0.1
epochs = 10000

# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Repeat for each epoch
for epoch in range(epochs):
    # Feed the training data forward through the network
    input_layer = X_train
    hidden_layer = sigmoid(np.dot(input_layer, W1) + b1)
    output_layer = sigmoid(np.dot(hidden_layer, W2) + b2)

    # Calculate the error between the predicted output and the target output
    error = y_train.reshape(len(y_train), 1) - output_layer

    # Backpropagate the error through the network to update the weights and biases
    d_output = error * sigmoid_derivative(output_layer)
    error_hidden_layer = d_output.dot(W2.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer)

    W2 += hidden_layer.T.dot(d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    W1 += input_layer.T.dot(d_hidden_layer) * learning_rate
    b1 += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

# Evaluate the performance of the network on the testing set
input_layer = X_test
hidden_layer = sigmoid(np.dot(input_layer, W1) + b1)
output_layer = sigmoid(np.dot(hidden_layer, W2) + b2)
predictions = np.round(output_layer)
accuracy = np.mean(predictions == y_test.reshape(len(y_test), 1))
print("Accuracy:", accuracy)
```
